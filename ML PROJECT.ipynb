## (ii) ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
training_model=date_group.iloc[:int(date_group.shape[0]*0.95)]
valid=date_group.iloc[int(date_group.shape[0]*0.95):]
log_series=np.log(training_model["Confirmed"])
pred_y=valid.copy()
```

```python
model_arima=ARIMA(log_series,(3,2,4))
arima_fit=model_arima.fit()
```

```python
pred_arima=arima_fit.forecast(len(valid))[0]
pred_y["ARIMA Model Prediction"]=list(np.exp(pred_arima))
```

```python
model_score.append([np.sqrt(mean_squared_error(list(valid["Confirmed"]),np.exp(pred_arima)))])
print("Root Mean Square Error for ARIMA Model: ",np.sqrt(mean_squared_error(list(valid["Confirmed"]),np.exp(pred_arima))))
```
